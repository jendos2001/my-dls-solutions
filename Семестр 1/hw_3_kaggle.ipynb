# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

Отнормируем признаки

```python
ss = StandardScaler()
ss.fit(data[num_cols])
data[num_cols] = ss.transform(data[num_cols])
```

```python
data.describe()
```

Применим OneHotEncoder

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(data[cat_cols])
data_copy = encoder.transform(data[cat_cols])
```

```python
text_columns = encoder.get_feature_names_out(cat_cols)
```

```python
data = data.drop(columns=cat_cols)
```

```python
data[text_columns] = data_copy
```

Разделим датасет и обучим его

```python
y = data[target_col]
X = data.drop(columns=[target_col])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, random_state=42)
lr = LogisticRegressionCV(refit=True, scoring='roc_auc')
```

```python
lr.fit(X_train, y_train)
```

```python
lr.get_params()
```

Рассчитаем метрики

```python
pred_train = lr.predict_proba(X_train)[:, 1]
pred_test = lr.predict_proba(X_test)[:, 1]
```

```python
roc_auc_score(y_train, pred_train)
```

```python
roc_auc_score(y_test, pred_test)
```

Применим GridSearch

```python
params = {
    'C': [100, 10, 1, 0.1, 0.01, 0.001],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 500, 1000, 5000]
}
```

Лучшие параметры и точность

```python
best_model1 = grid.best_estimator_
```
